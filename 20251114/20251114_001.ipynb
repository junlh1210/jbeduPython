## Dictionary 딕셔너리
중괄호를 이용하여 데이터를 표현  
{key1:value1, key2:value2, ... }

```python
dic1={1:'a', 2:'b', 3:'c'}
dic2={'a':100, 'b':200, 'c':300}

print(dic1)
print(dic1[1])      ##  dic1의 key가 1 인 value 
print(dic1[3])
print(dic2)
print(dic2['a'])    ##  dic2의 key가 a 인 value 
print(dic2['c'])
print(dic2['a'] + dic2['c'])

print(dic1.keys())
print(dic1.values())
print(dic1.items())
```

```python
member = {"kim":"danwoong", "hong":"hyeji", "park":"miso"}
print(member)
```

```python
# 디셔너리 요소에 접근
member = {"kim":"danwoong", "hong":"hyeji", "park":"miso"}
print(member['park'])
```

```python
member = {"kim":"danwoong", "hong":"hyeji", "park":"miso"}
member['park']='smile'
member['lee']='danbi'
print(member)
```

```python
member = {"kim":"danwoong", "hong":"hyeji", "park":"miso"}
del member['park']
print(member)
```

```python
eng = {'break':'깨지다','bring':'가져오다','brush':'붓','brush':'빗','brave':'용감한'}
print(eng['break'])
print(eng['brush'])     # 중복 key라면 마지막값이 출력

print(eng['brief'])   ## 에러
```

```python
eng = {'break':'깨지다','bring':'가져오다','brush':['붓', '빗'],'brave':'용감한'}
print(eng['brush'])
print(eng['brush'][0])
print(eng['brush'][1])
a = input('단어입력 : ')    # eng에 있는 key를 입력한다
print(eng[a])
```

```python
menu = {'김밥':2500, '참치김밥': 3500, '라면':3000, '치즈라면': 4000}
print(menu)

# 돈까스 6000원 추가
menu['돈까스'] = 6000
print(menu)

# 깁밥은 3000원으로 수정
menu['김밥'] = 3000
print(menu)

# 라면의 가격은 200원 인하
menu['라면'] = menu['라면']-200
print(menu)

## 치즈라면 삭제
del menu['치즈라면']
print(menu)
```

```python
menu = {'김밥':2500, '참치김밥': 3500, '라면':3000, '치즈라면': 4000}

print(sorted(menu.keys()))
print(sorted(menu.keys(), reverse=True))

print(sorted(menu.values()))
print(sorted(menu.values(), reverse=True))

print(sorted(menu.items()))
print(sorted(menu.items(), reverse=True))

menu.sort()     # 딕셔너리는 sort를 사용할 수 없다
```
