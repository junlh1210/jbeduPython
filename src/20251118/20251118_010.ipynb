```python
import matplotlib.pyplot as plt
# plt.rc('font',family='Malgun Gothic')
# plt.rcParams['axes.unicode_minus']=False  # '- 표시

plt.rc('font',family='D2CodingLigature Nerd Font')
```

```python
month=['mar', 'apr','may','jun','jul']
sales=[1,5,7,3,7]

plt.title('월별 판매 실적')
plt.plot(month,sales, color='b', label='단위: 개수') #범례넣기
plt.legend()
plt.show()
```

```python
month=['mar', 'apr','may','jun','jul']
sales=[1,5,7,3,7]

plt.title('월별 판매 실적')
plt.bar(month,sales, color='y')
plt.show()
```

```python
month=['mar', 'apr','may','jun','jul']
sales=[1,5,7,3,7]

plt.figure(figsize=(5,2))
plt.title('월별 판매 실적')
plt.bar(month,sales, color='y')
plt.show()
```

```python
b_type=[25,19,37,11]
b_name=['A형','B형','O형','AB형']

plt.pie(b_type, labels=b_name, autopct='%.1f%%')
plt.title('혈액형 비율')

plt.show()
```

```python
b_type=[25,19,37,11]
b_name=['A형','B형','O형','AB형']
exp=[0,0,0.1,0]

plt.pie(b_type, labels=b_name, autopct='%.1f%%', explode=exp)
plt.title('혈액형 비율')

plt.show()
```

```python
kor=[80,20,50,20,10,50,60,30,60]
eng=[90,40,60,40,10,30,50,70,90]

plt.scatter(kor,eng,color='r',alpha=0.3)

plt.title('국어와 영어 점수의 상관관계')
plt.xlabel('국어')
plt.ylabel('영어')
plt.show()
```

```python
singer=['A','B','C','D','E']
week1=[42,58,19,92,84]
week2=[53,52,48,98,73]

plt.plot(singer,week1, label="첫째 주", color="hotpink")
plt.plot(singer,week2, label="둘째 주", color="royalblue")
plt.legend()
plt.show()
```

```python
singer=['A','B','C','D','E']
week1=[42,58,19,92,84]
week2=[53,52,48,98,73]

plt.title('오디션 프로그램 득표 현황')

plt.bar(singer,week1,label='첫째 주', color='hotpink')
plt.bar(singer,week2, label='둘째 주', color='royalblue')
plt.legend()
plt.show()
```

```python
# 쥬피터노트에서 -기호 사용시
# plt.rcParams['axes.unicode_minus']=False  # '- 표시

singer=['A','B','C','D','E']
week1=[42,58,19,92,84]
week2=[53,52,48,98,73]

for i in range(len(week1)):
    week1[i] *= -1

plt.title('오디션 프로그램 득표 현황')

plt.barh(singer,week1,label='첫째 주', color='hotpink')
plt.barh(singer,week2, label='둘째 주', color='royalblue')
plt.legend()
plt.show()
```
