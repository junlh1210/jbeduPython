## 튜플(tuple)

파이썬에서의 자료구조  
수정 불가한 데이터 구조를 말한다

```python
tt1=(10,20,30)
tt2=10,20,30                
print(type(tt1))            # <class 'tuple'>
print(type(tt2))            # <class 'tuple'>
print(tt1)                  # (10, 20, 30)
print(tt2)                  # (10, 20, 30)
print(tt1[0])               # 10
print(tt1[0]+tt1[1]+tt1[2]) # 60
print(tt1[:1])
print(tt1[1:])
      
tt1.append(40)  ## 에러
tt1
```
