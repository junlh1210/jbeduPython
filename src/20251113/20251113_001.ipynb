```python
## 입력받은 인원수만큼 점수를 입력받아,
## 리스트에 점수를 추가하는 코드 작성


list1 =[]
a = int(input("인원수 : "))

for  i  in range(a):
   x = int(input("점수 :" ))
   list1.append(x)

print(list1)
```

```python
# 비어있는 리스트를 생성하고,
# 인원수를 입력받아, 사용자로부터 인원수만큼 반복하면서
# '이름'과 '나이'를 입력 받는다.
# 입력받은 이름과 나이를 비어있는 리스트에 한쌍으로 추가한다.
# 입력이 마무리되면, 입력받은 정보를 출력한다.

list1 = []
num = int(input("인원수 : "))
for i in range(num):
    x = input("이름 : ")
    y = int(input("나이 : "))
    list1.append([x, y])

print("===입력된 사용자 정보===")
for x, y in list1:
    print(f"이름 : {x}\t나이 : {y}")
```

```python
## 비어있는 리스트를 생성하고,
## 상품 개수를 입력받아, 입력받은 개수만큼 반복하면서
## 상품이름과 상품가격을 입력 받는다.
## 입력받은 이름과 가격을 비어있는 리스트에 한쌍으로 추가한다.
## 입력이 마무리되면, 모든 상품의 정보를 출력한다.
## 총 가격과 평균 가격을 계산하여 출력한다.

list1 = []
item_num = int(input("상품수 : "))
for i in range(item_num):
    item_name = input("상품이름 : ")
    item_price = float(input("상품가격 : "))
    list1.append([item_name, item_price])
print(list1)

total = 0
for i in list1:
    total += i[1]
print(f"총가격 : {total}\t평균가격 : {total/item_num:.2f}")
```

```python
## 비어있는 리스트를 생성하고,
## 상품 개수를 입력받아, 입력받은 개수만큼 반복하면서
## 상품의 이름과 가격, 할인율을 입력받는다.
## 입력받은 이름과 가격, 할인율을 어있는 리스트에 한쌍으로 추가한다.

## 모두 입력되면 다음과 같은 정보를 출력한다.
## 상품이름, 가격, 할인율, 할인 적용된가격

products = []
num = int(input("상품 수 : "))

for i in range(num):
    products_name = input("상품 이름 : ")
    products_price = float(input("상품 가격 : "))
    products_discount_rate = float(input("할인율 : "))
    products_sale_price = products_price * (100-products_discount_rate)/100
    products.append([products_name, products_price, products_discount_rate, products_sale_price])

print("\n*** 상품별 할인 정보 ***")
for n, p, d, s in products:
    print(f"{n:10s}\t | 원가: {p:10,.2f} | 할인율: {d:5.1f}% | 판매가: {s:10,.2f}")
```
