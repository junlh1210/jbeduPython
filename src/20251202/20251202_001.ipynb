## 결측치 처리 _결측치 제거

```python
# 결측치 있는 행 지우기 (axis=1 이면 열을 지운다)

air_d = air.dropna(axis=0)
air_d
```

```python
air_d.reset_index(inplace=True)
air_d
```

## 결측치 처리 _평균값으로 대체

```python
# 결측치 평균대체를 위해 평균 확인

mean_air = air.mean(numeric_only=True)
mean_air
```

```python
# 평균으로 결측치를 채우기

air_m = air.fillna(mean_air)
air_m
```

## 중앙값으로 채우기

```python
# 중앙값 구하기

air_meadian = air.median(numeric_only=True)
air_meadian
```

```python
# 중앙값으로 결측치 채우기

air_md = air.fillna(air_meadian)
air_md
```
