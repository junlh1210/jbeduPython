## Titanic.csv를 이용한 데이터 전처리

```python
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

```python
titanic_df= pd.read_csv("../../data/titanic.csv")
titanic_df.info()
```

```python
titanic_df.isna().sum()
```

#### age의 결측치는 평균으로 대체

```python
# age의 평균 

titanic_df['age'].mean()
```

```python
# 특정행 및 열의 자료 보여주기

titanic_df['age'][15:20]
```

```python
titanic_df['age'] = titanic_df['age'].fillna(titanic_df['age'].mean())
titanic_df['age'][15:20]
```

```python
titanic_df.isna().sum()
```

#### embark_town 결측치는 최빈값으로 대체

```python
titanic_df['embark_town'][825:830]
```

```python
# value_counts() : 유니크 값이 몇 번나오는지 새어준다

most_freq = titanic_df['embark_town'].value_counts(dropna=True)
most_freq
```

```python
# embark_town 열의 최빈값
most_freq = titanic_df['embark_town'].value_counts(dropna=True).idxmax()  # 최대값을 가진 인덱스
most_freq
```

```python
titanic_df['embark_town'] = titanic_df['embark_town'].fillna(most_freq)
```

```python
titanic_df['embark_town'][825:830]
```

```python
# embark_town 결측치 대체 확인

titanic_df.info()
```
