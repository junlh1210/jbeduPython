## where 함수
    * 조건이 참인지 거짓인지에 따라 다른 값을 선택하여 반환
    * 배열의 모든 요소를 처리

```python
import warnings
warnings.filterwarnings(action='ignore') 

import numpy as np
import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

```python
score=[71,80,60,90,65]
result=['pass' if a>=70 else 'fail'  for a in score]
print(result)
```

```python
score=[71,80,60,90,65]
result = []
for i in score:
    if i>=90: result.append("A")
    elif i>=80 : result.append("B")
    elif i>=70 : result.append("C")
    elif i>=60 : result.append("D")
    else: result.append("F")

print(result)
```

```python
score=np.array([71,80,60,90,65])
result=np.where(score>=90, 'A', np.where(score>=80, "B", np.where(score>=70, "C", np.where(score>=60,"D","F"))))
print(result)
```

## 출생아수 그래프 그리기

```python
year = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
boy = [248958,223883,223356,224906,208064,184308,167686,155416,139362,133516]
girl = [235592,212572,212079,213514,198179,173463,159136,147260,132975,127046]
```

#### 1. 남아출생아수와 여아출생아수 막대그래프 그리기

```python
ind = np.arange(len(year))

plt.figure(figsize=(6, 3.5))
plt.bar(ind-0.2, boy, color='b', width=0.4, label='boy')
plt.bar(ind+0.2, girl, color='r', width=0.4, label='girl')
plt.xticks(ind, year)
plt.legend()
plt.xlabel('year')
plt.ylabel('number of births')
plt.show()
```

#### 2. 신생아수 누적 막대그래프 그리기

```python
ind = np.arange(len(year))

plt.figure(figsize=(6, 3.5))
plt.bar(ind, boy, color='b', width=0.4, label='boy')
plt.bar(ind, girl, bottom=boy, color='r', width=0.4, label='girl')
plt.xticks(ind, year)
plt.legend()
plt.xlabel('year')
plt.ylabel('number of births')
plt.show()
```
