```python
import numpy as np 
a=np.arange(1,10,2)
b=np.arange(2,11,2)
c=np.arange(-1,-11,-2)
d=np.arange(1,4)    # [1 2 3]

print(f"a : {a}")
print(f"b : {b}")
print(f"c : {c}")
print(f"d : {d}")
print()

print(f"np.add(a,b) : {np.add(a,b)}")
print(f"np.subtract(b,1) : {np.subtract(b,1)}")
print(f"np.multiply(a,b) : {np.multiply(a, b)}")
print(f"np.divide(a,b) : {np.divide(a,b)}")
print(f"np.floor_divide(a,2) : {np.floor_divide(a,2)}")  ## 몫
print(f"np.mod(a,2) : {np.mod(a,2)}")    ## 나머지
print(f"np.power(a,3) : {np.power(a, 3)}")  ## 3제곱
print(f"np.abs(c) : {np.abs(c)}")   ## 절대값
print(f"np.sqrt(b) : {np.sqrt(b)}")  ## 루트
print(f"np.square(a) : {np.square(a)}")  ## 제곱
print(f"np.exp(d) : {np.exp(d)}")   ## 자연상수 2.718의 d승 
```

## 넘파이의 통계함수

```python
a=np.array([3,7,4,11,5])
print(np.sum(a))        # 합
print(np.mean(a))       # 평균
print(np.std(a))        # 표준편차
print(np.var(a))        # 분산
print(np.max(a))        # 최대값
print(np.min(a))        # 최소값
print(np.argmax(a))     # 최대값의 인덱스
print(np.argmin(a))     # 최소값의 인덱스
print(np.median(a))     # 중앙값
print(np.cumsum(a))     # 누적 합
print(np.cumprod(a))    # 누적 곱 
```

```python
math = np.array([75, 89, 75, 65, 79])
english = np.array([85, 80, 90, 75, 88])
korean = np.array([72, 90, 70, 88, 93])
x = ['math', 'english', 'korean']

plt.figure(figsize=(5, 3))
plt.bar(x, [np.mean(math), np.mean(english), np.mean(korean)], color=['r', 'g', 'b'])
plt.ylabel('score')
plt.show()
```
