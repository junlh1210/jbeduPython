```python
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

```python
df = pd.DataFrame({'name': ['Jessica', 'Liam', 'Sophia', 'Ryan', 'Alex'],
                   'test': [45, 30, 40, 37, 48],
                   'assign1': [20, 17, 22, 18, 24],
                   'assign2': [19, 14, 18, 15, 25],
                   'total': [84, 61, 80, 70, 97]})

df
```

```python
df.rename(columns={'test':'exam'})   ## 열 이름 변경이 적용되지 않음
df
```

```python
df2 = df.rename(columns={'test':'exam'})    # 원본데이터프레임이 아닌 새로운 데이터프레임 생성
df2
```

```python
df.rename(columns={'test':'exam'}, inplace=True)    ## 원본의 열 이름 변경 적용됨
df
```

```python
df['grade'] = np.where(df['total']>=90, 'A', np.where(df['total']>=80, 'B', 'C'))   # numpy의 where 함수
df
```

```python
df['result'] = np.where(df['total']>=80, 'PASS', 'FAIL')
df
```

```python
df['result'].value_counts()     # 요소를 구분하여 count
```

```python
df['grade'].value_counts()
```

```python
df[['grade', 'result']].value_counts()      # 3X2=6 총 6개의 경우의 수가 있을 수 있다
```
