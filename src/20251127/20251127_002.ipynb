```python
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

```python
df1 = pd.DataFrame({'name': ['Jessica', 'Liam', 'Sophia', 'Ryan','Alex'],
                    'gender': ['F', 'M', 'F', 'M', 'F'],
                    'age': [21, 23, 20, 22, 20]})

df2 = pd.DataFrame({'name': ['Sophia', 'Liam', 'Ryan', 'Alex', 'Jessica'],
                    'department': ['computer science', 'mathematics', 'law',
                    'computer science', 'law']})
```

```python
display(df1)
display(df2)
```

```python
df3 = pd.merge(df1, df2, on='name')     # inner 방식
df3
```

```python
df4 = pd.merge(df1, df2, on='name', how='outer')    # outer 방식
df4

## how = 'left'     df1 기준
## how = 'right'    df2 기준
```

```python
df5 = pd.DataFrame({'department': ['computer science', 'mathematics', 'law'],
                    'head of department': ['Steve', 'Emma', 'Carly']})
df5
```

```python
df6 = pd.merge(df4, df5, on='department')
df6
```

```python
df7 = df6.sort_values(['department', 'age'], ascending=False)
df7
```
