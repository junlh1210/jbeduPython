### 문자열 함수

```python
df = pd.DataFrame({"높이":['1,250미터','740','높이1,130미터']})
df
```

```python
df['높이'].str.find('높')   ## 지정한 문자열을 찾아서, 그 인덱스 반환, 해당 문자열이 없으면 -1
```

```python
df['높이'].str.find('0')
```

```python
df['높이'] = df['높이'].str.strip()   ## 문자열 양 끝의 공백 또는 해당 문자 제거
df
```

```python
df['높이'] = df['높이'].str.rstrip('미터')    ## 오른쪽 공백 또는 해당 문자('미터') 제거
df
```

```python
df['높이'] = df['높이'].str.lstrip('높이')    ## 왼쪽 공백 또는 해당 문자 제거
df
```

```python
df['높이'] = df['높이'].str.replace(',','')    ## 해당 문자 교체 (컴마를 제거)
df
```

```python
df.info()
```

```python
# 문자열 -> 숫자

df['높이'] = df['높이'].astype(int)
df.info()
```
