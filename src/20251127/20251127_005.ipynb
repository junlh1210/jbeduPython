```python
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

#### 국가별 메달획득 그래프

```python
df = pd.DataFrame({'nation': ['Korea.', 'USA.', 'Japan', 'France'],
                   'gold': [12, 36, 4, 6],
                   'silver': [10, 31, 3, 4],
                   'bronze': [11, 27, 7, 6]})
display(df)
```

```python
df.info()
```

```python
# 데이터 앞뒤의 공백 제거

df['nation'] = df['nation'].str.strip()
df
```

```python
# nation에서 마지막의 마침표 제거

df['nation'] = df['nation'].str.replace(".",'')
df
```

```python
plt.figure(figsize=(5,3))
plt.bar(df['nation'], df['gold'], color='gold', label='gold')
plt.bar(df['nation'], df['silver'], color='silver', bottom=df['gold'], label='silver')
plt.bar (df['nation'], df['bronze'], color='darkorange', bottom=df['gold']+df['silver'], label='bronze')
plt.legend()
plt.title('1988 서울 올림픽 국가별 획득 메달 수')
plt.ylabel('number of medals')
plt.show()
```
