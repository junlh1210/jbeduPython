### 통계함수

```python
df = pd.DataFrame({'test': [45,30,40,37,48]})
df.count()
```

```python
df = pd.DataFrame({'test': [45,30,40,37,48], 'test1':[10, 20, 30, np.nan, 100]})
df.count()      # 결측값 제외
```

```python
df.sum()
```

```python
df.cumsum()     # 누적합
```

```python
df.mean()       # 각각 열의 평균
```

```python
df.std().round(2)       # 표준편차, 소숫점 2자리까지
```

```python
df.var()    # 분산
```

```python
df.max()    # 최대값
```

```python
df.min()    # 최소값
```

```python
df.describe()   # 알아서 특정값들을 보여준다
```
