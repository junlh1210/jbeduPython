## 덴드로그램(dendrogram)

```python
f = np.array([[3, 4], [1, 5], [7, 9],[5, 4], [6, 8], [2, 3],[4,4]])
plt.figure(figsize=(20,10))
cluster=linkage(f,method='ward') 
## ward : 클러스터 내의 분산을 최소화하는 방향으로 클러스터 간의 거리를 계산

dendrogram(cluster)
plt.title('Dendrogram')
plt.xlabel('Sour taste')
plt.ylabel('sweet taste')
```

```python
f = np.array([[3, 4], [1, 5], [7, 9],[5, 4], [6, 8], [2, 3],[4,4]])
plt.figure(figsize=(20,10))
cluster=linkage(f,method='single') 
## ward : 클러스터 내의 분산을 최소화하는 방향으로 클러스터 간의 거리를 계산
## single : 가장 가까운 데이터들
## complete : 가장 먼 데이터들 
## centroid : 중심점간의 거리 
## median : 클러스터 간의 중앙값 기준으로 병합

dendrogram(cluster)
plt.title('Dendrogram')
plt.xlabel('Sour taste')
plt.ylabel('sweet taste')
```

```python
f = np.array([[3, 4], [1, 5], [7, 9],[5, 4], [6, 8], [2, 3],[4,4]])
plt.figure(figsize=(20,10))
cluster=linkage(f,method='average') 
## average : 평균 연결 방법
## 각 클러스터의 데이터들 간 모든 거리들의 평균을 클러스터 간 거리로 사용

dendrogram(cluster)
plt.title('Dendrogram')
plt.xlabel('Sour taste')
plt.ylabel('sweet taste')
```
