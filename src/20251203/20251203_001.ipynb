## loc와 iloc (조건을 붙이려면 loc)

```python
data = {
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
}
df = pd.DataFrame(data, index=['a','b','c','d'])
df
```

```python
df.loc[df['A']>=3]
```

```python
df.loc[df['A']>=3, "C"]
```

```python
df.iloc[1, 0]
```

```python
df.loc['b','A']
```

```python
# df.loc['b':'d', 'A':'B']

df.iloc[1:4, 0:2]   # 숫자는 레인지처럼 계산된다
```

```python
df.loc[df['A']>=2, 'A':'B']
```

```python
# df.loc[['a','c'], ['A','C']]

df.iloc[ [0,2] , [0,2] ]
```
