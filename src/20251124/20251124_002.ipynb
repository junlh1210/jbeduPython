## 서울생활인구_남녀_지도표시

```python
import warnings
warnings.filterwarnings(action='ignore') 

import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

#### 남성 인구수를 기준으로 내림차순 정렬하여 위의 10개 데이터만 지도에 표시하기

```python
# 인구 데이터 읽어와서 리스트로 저장하기
f1 = open('../../data/LOCAL_PEOPLE_DONG_202510.csv', encoding='utf8')    
data = csv.reader(f1)
next(data)
data = list(data)

# 행정동코드 데이터 불러와서 리스트로 저장하기
f2 = open('../../data/dong_code.csv')
code_data = csv.reader(f2)
next(code_data)
next(code_data)
code_data = list(code_data)

# 위치 데이터 불러와서 리스트로 저장하기
f3 = open('../../data/dong_location.csv')
location_data = csv.reader(f3)
next(location_data)
location_data = list(location_data)

# 인구데이터의 데이터타입 변경하기 
for row in data : 
    for i in range(1, 32) : 
        if i <= 2 : 
            row[i] = int(row[i])
        else : 
            row[i] = float(row[i])
            
# 행정동코드 데이터의 데이터 타입 변환하기
for row in code_data :
    row[1] = int(row[1]) # 8자리 행정동코드

# 위치 정보 데이터의 데이터 타입 변환하기
for row in location_data :
    row[-2], row[-1] = float(row[-2]), float(row[-1])  ### row[-2]:경도, row[-1]:위도
```

```python
# 행정동별 남성 인구수의 총합을 저장하기 위한 딕셔너리 준비
dong_population_m = {}

for row in data: 
    dong_code, p = row[2], sum(row[4:18])    ## 4~17번 인덱스 값이 남성 데이터
    if dong_code not in dong_population_m.keys(): 
        dong_population_m[dong_code] = p
    else:
        dong_population_m[dong_code] += p

print(dong_population_m) 
```

```python
# 남성 인구수를 기준으로 내림차순 정렬하여 위 10개만 선택
import operator

top10_m = sorted(dong_population_m.items(), key=operator.itemgetter(1), reverse = True)[:10]
print(top10_m)
print("")

top10_code8_m = [t[0] for t in top10_m]
print(top10_code8_m)
```

```python
top10_name_m = [0 for i in range(10)]
top10_lat_m  = [0 for i in range(10)]
top10_long_m = [0 for i in range(10)]

# 행정동 코드에 맞는 행정동명 찾기

for i in range(len(top10_code8_m)) : 
    for row in code_data : 
        if row[1] == top10_code8_m[i] :  
            top10_name_m[i] = row[-1]      # 행정동명 리스트에 저장 

    for row in location_data : 
        if row[3] == top10_name_m[i] :
            top10_lat_m[i], top10_long_m[i] = row[-1], row[-2]  # 행정동명의 위도와 경도를 찾아 리스트에 저장

print(top10_name_m)
print(top10_lat_m)
print(top10_long_m)
```

```python
top10_name_mr =[0   for i in range(10)]

for i in range(len(top10_name_m)) : 
    top10_name_mr[i] = str(i+1)+'. '+top10_name_m[i]

map_top10_m = folium.Map([top10_lat_m[0], top10_long_m[0]], zoom_start=11) 
for i in range(10):
    if i < 5: 
        folium.Marker(
            [top10_lat_m[i], top10_long_m[i]], tooltip=top10_name_mr[i], 
            icon=folium.Icon(color='darkblue', icon='male', prefix='fa')
        ).add_to(map_top10_m)
    else: 
        folium.Marker(
            [top10_lat_m[i], top10_long_m[i]], tooltip=top10_name_mr[i], 
            icon=folium.Icon(color='lightblue', icon='male', prefix='fa') 
        ).add_to(map_top10_m)

map_top10_m
```

#### 여성 인구수를 기준으로 내림차순 정렬하여 위의 10개 데이터만 지도에 표시하기

```python
dong_population_f = {}    # 행정동별 여성 인구수를 저장하는 딕셔너리 

for row in data: 
    # 여성 인구수를 저장하는 변수 p
    dong_code, p = row[2], sum(row[18:32])
    if dong_code not in dong_population_f.keys(): 
        dong_population_f[dong_code] = p
    else:
        dong_population_f[dong_code] += p
        
# print(dong_population_f) 

top10_f = sorted(dong_population_f.items(), key=operator.itemgetter(1), reverse = True)[:10]
#print(top10_f)

top10_code8_f = [t[0] for t in top10_f]
print(top10_code8_f)


top10_name_f = [0 for i in range(10)]
top10_lat_f  = [0 for i in range(10)]
top10_long_f = [0 for i in range(10)]

# 4. 8자리 행정동코드에 해당하는 행정동명 찾기
for i in range(len(top10_code8_f)) : 
    for row in code_data : 
        if row[1] == top10_code8_f[i] :  
            top10_name_f[i] = row[-1]

    for row in location_data : 
        if row[3] == top10_name_f[i] :
            top10_lat_f[i], top10_long_f[i] = row[-1], row[-2] 

print(top10_lat_f)
print(top10_long_f)


top10_name_fr =[0   for i in range(10)]

for i in range(len(top10_name_f)) : 
    top10_name_fr[i] = str(i+1)+'. '+top10_name_f[i]

    
map_top10_f = folium.Map([top10_lat_f[0], top10_long_f[0]], zoom_start=11) 
for i in range(10):
    if i < 5: 
        folium.Marker(
            [top10_lat_f[i], top10_long_f[i]], tooltip=top10_name_fr[i], 
            icon=folium.Icon(color='red', icon='female', prefix='fa')
        ).add_to(map_top10_f)
    else: 
        folium.Marker(
            [top10_lat_f[i], top10_long_f[i]], tooltip=top10_name_fr[i], 
            icon=folium.Icon(color='pink', icon='female', prefix='fa') 
        ).add_to(map_top10_f)

map_top10_f
```
