## groupby() 와 agg()

#### DataFrame.groupby(기준항목).agg(사용하고자 하는 function들)

```python
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

```python
# pandas에서 csv파일 읽어오기

exam = pd.read_csv('../../data/exam.csv')
exam
```

```python
exam.groupby('nclass')[['math', 'english']].mean()
```

```python
exam.groupby('nclass').mean()
```

```python
exam.groupby('nclass').agg(mean_math = ('math', 'mean'), mean_english = ('english', 'mean'))
```

```python
exam.groupby('nclass').agg(mean_math = ('math', 'mean'), 
                           sum_english = ('english', 'sum'),
                           median_math = ('math', 'median'), 
                           n = ('nclass','count')
                          )
```
