## csv 읽기 및 쓰기

```python
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

```python
# pandas에서 csv파일 읽어오기

df = pd.read_csv('../../data/precipitation.csv')
df
```

```python
plt.figure(figsize=(6,4))
plt.bar(df['year'], df['spring'], label='spring')
plt.bar(df['year'], df['summer'], bottom=df['spring'], label='summer')
plt.bar(df['year'], df['autumn'], bottom=(df['spring']+df['summer']), label='autumn')
plt.bar(df['year'], df['winter'], bottom=(df['spring']+df['summer']+df['autumn']), label='winter')
plt.title('년도별 계절별 강수량')
plt.ylabel('precipitation')
plt.xlabel('year')
plt.legend()
plt.show()
```

```python
df['total'] = df['spring']+df['summer']+df['autumn']+df['winter']
df
```

```python
# pandas에서 csv파일 쓰기

df.to_csv('../../data/precipitation_out1.csv')
```
