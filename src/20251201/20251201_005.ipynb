```python
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
import csv
import folium
import matplotlib.pyplot as plt
plt.rc('font',family='D2CodingLigature Nerd Font')
```

```python
# pandas에서 csv파일 읽어오기

df = pd.read_csv('../../data/class.csv')
df
```

```python
ind = np.arange(len(df))
plt.figure(figsize=(5,3))
plt.plot(ind, df['서울특별시'], label='seoul')
plt.plot(ind, df['제주특별자치도'], label='jeju')
plt.xticks(ind, df['시점'])
plt.ylim(15, 30)
plt.legend()
plt.xlabel('년도', fontsize=15)
plt.ylabel('학습당 학생 수', fontsize=10)
plt.xticks(rotation='vertical')
plt.show()
```
