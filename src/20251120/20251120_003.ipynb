```python
import csv      # 기본 모듈
import folium

f1 = open('../../data/seoul_uni_location.csv', encoding='utf8')
loc_data=csv.reader(f1)
next(loc_data)              # 1행 제거
loc_data=list(loc_data)     # list형태로 변환

map1 = folium.Map([37.56678, 126.9782], zoom_start=12)    # 서울시청을 중심으로

for uni in loc_data:
    folium.Marker([uni[1],uni[2]],tooltip=uni[0]).add_to(map1)

map1.save("map01_대학위치.html")      # 지도를 html문서로 저장
map1
```
