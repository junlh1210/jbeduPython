```python
import csv      # 기본 모듈
import folium

f1 = open('../../data/dong_location.csv', encoding='utf8')
loc_data=csv.reader(f1)
next(loc_data)              # 1행 제거
loc_data=list(loc_data)     # list형태로 변환

dong_name = input("행정동명을 입력하세요 : ")

for r in loc_data:
    if r[3] == dong_name:
        a = r[-1]
        b = r[-2]

map1 = folium.Map([a,b], zoom_start=15)
folium.Marker([a,b],tooltip=dong_name).add_to(map1)
map1.save("map01_행정동.html")      # 지도를 html문서로 저장
map1
```

```python
import csv      # 기본 모듈
import folium

f1 = open('../../data/dong_location.csv', encoding='utf8')
loc_data=csv.reader(f1)
next(loc_data)              # 1행 제거
loc_data=list(loc_data)     # list형태로 변환

dong_name1 = input("첫번째 행정동명을 입력하세요 : ")
dong_name2 = input("두번째 행정동명을 입력하세요 : ")

for r in loc_data:
    if r[3] == dong_name1:
        a = r[-1]
        b = r[-2]
    elif r[3] == dong_name2:
        c = r[-1]
        d = r[-2]

map1 = folium.Map([a,b], zoom_start=15)
folium.Marker([a,b],tooltip=dong_name1).add_to(map1)
folium.Marker([c,d],tooltip=dong_name2).add_to(map1)
map1.save("map02_행정동.html")      # 지도를 html문서로 저장
map1
```
