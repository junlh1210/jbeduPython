### 다양한 마커 아이콘 : https://fontawesome.com/v4.7.0/icons

```python
# 중부기술교육원    37.54025, 127.0048
# 서울시청          37.56678, 126.9782
# 스페인 대사관     37.54191, 127.0045

import folium   # 지도 라이브러리

map_y = folium.Map([37.56678, 126.9782], zoom_start=15)    # 서울시청을 중심으로 Map을 얻어5

folium.Marker([37.54025, 127.0048]
              , tooltip="서울시 중부기술 교육원"
              , icon=folium.Icon(color='pink', icon='automobile', prefix='fa')).add_to(map_y)       # 중부기술교육원 마크
folium.Marker([37.54191, 127.0045]
              , tooltip="스페인 대사관"
              , icon=folium.Icon(color='blue', icon='asterisk', prefix='fa')).add_to(map_y)         # 스페인 대사관

map_y   # 지도 보이기
```

```python
## 여의도 공원       37.52648, 126.9229
## 여의도 한강공원   37.52692, 126.9347
## 샛강 생태공원     37.51890, 126.9219

a,b = 37.52648, 126.9229    # 위도, 경도값을 나누어서 저장
c,d = 37.52692, 126.9347
e,f = 37.51890, 126.9219

map1 = folium.Map([a,b], zoom_start=15)     # 지도 중심과 크기 지정

folium.Marker([a,b], tooltip='여의도 공원', icon=folium.Icon(color='pink', icon='automobile', prefix='fa')).add_to(map1)
folium.Marker([c,d], popup='여의도 한강공원').add_to(map1)  # 마우스를 클릭하여야 보인다
folium.Marker([e,f], tooltip='여의도 샛강생태공원', icon=folium.Icon(color='orange', icon='ban', prefix='fa')).add_to(map1)

map1
```

```python
a = [37.52648, 37.52692, 37.51890]      # 위도 리스트
b = [126.9229, 126.9347, 126.9219]      # 경도 리스트
c = ['여의도 공원', '여의도 한강공원', '여의도 샛강생태공원']   # 툴팁 리스트
d = ['automobile', 'balance-scale', 'ban']     # 아이콘 모양 리스트
f = ['red', 'blue', 'purple']    # 아이콘 색상 리스트

map2 = folium.Map([a[0], b[0]], zoom_start = 15)

for i in range(len(a)) : 
    folium.Marker([a[i], b[i]]
                  , tooltip=c[i]
                  , icon=folium.Icon(color=f[i], icon=d[i], prefix='fa')).add_to(map2)

map2
```
