## 카드사용내역 데이터

```python
import csv

f=open('../../data/card.csv', encoding='utf8')
data=csv.reader(f)
next(data)  # 열 이름 제거
```

```python
data=list(data)
print(data)
```

```python
for i in data:
    print(i[0], i[5], i[6])
```

```python
s = 0
for r in data:
    s = s + int(r[6])

print(f"총 {s:,}원")    # 이용금액 총계
```

```python
## "전표매입"으로 표시되어 있는 행과 그렇지 않은 행으로 나누어서 
## "구매확정!"과 "구매취소!"로 표현하기


for row in data:
    if row[-1]=='전표매입':
        str1 = "구매확정!"
    else:
        str1 = "구매취소!"
    print(f"{row[0]} : {row[5]}에서 {int(row[6]):,}원 결제 ==> {str1}")
```

```python
## '전표매입'된 승인 금액의 총합

for row in data:
    if row[-1]=='전표매입' :
        s = s + int(row[-3])
        
print(f"승인 총 금액 : {s:,}") 
```

```python
## '전표매입'이 아닌 행의 금액 총합

s = 0
for row in data:
    if row[-1]!='전표매입' :
        s = s + int(row[-3])
        
print(f"승인 총 금액 : {s:,}") 
```

```python
## 10월 승인 금액의 합계

mon10=0
for row in data:
    if row[-1]=='전표매입' and int(row[0].split('-')[1])==10:   # '-'을 기준으로 분리
        mon10+=int(row[-3])
        
print(mon10) 
```

```python
mon10, mon11, mon12 = 0,0,0

for row in data:
    if row[-1]=='전표매입':
        if int(row[0].split('-')[1])==10:
            mon10 = mon10 + int(row[-3])
        elif int(row[0].split('-')[1])==11:
            mon11 = mon11 + int(row[-3])
        elif int(row[0].split('-')[1])==12:
            mon12 = mon12 + int(row[-3])
        
print(f"10월 승인 총 금액 : {mon10:,}원") 
print(f"11월 승인 총 금액 : {mon11:,}원") 
print(f"12월 승인 총 금액 : {mon12:,}원") 
```

```python
## 10, 11, 12월 승인 총 금액

s_mon = [0, 0, 0]

for row in data:
    if(row[-1]=='전표매입'):
        if(int(row[0].split('-')[1])==10):
            s_mon[0] = s_mon[0] + int(row[-3])
        elif (int(row[0].split('-')[1])==11):
            s_mon[1] = s_mon[1] + int(row[-3])
        elif (int(row[0].split('-')[1])==12):
            s_mon[2] = s_mon[2] + int(row[-3])
        
print(f"10월 승인 총 금액 : {s_mon[0]:,}원") 
print(f"11월 승인 총 금액 : {s_mon[1]:,}원") 
print(f"12월 승인 총 금액 : {s_mon[2]:,}원") 
```

```python
## 10, 11, 12월 승인 총 금액

s_mon = [0, 0, 0]

for row in data:
    if(row[-1]=='전표매입'):
        m = int(row[0].split('-')[1])
        s_mon[m-10] += int(row[-3])

for i in range(10, 12+1,1):
    print(f"{i}월 승인 총 금액 : {s_mon[i-10]:,}원") 
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family="D2CodingLigature Nerd Font")

s_mon = [0, 0, 0]

for row in data:
    if(row[-1]=='전표매입'):
        m = int(row[0].split('-')[1])
        s_mon[m-10] = s_mon[m-10] + int(row[-3])

plt.title("10-12월 지출현황")
plt.bar(['10월', '11월', '12월'], s_mon, color='royalblue')
plt.show()
```

```python
## "전표매입"데이터 중 "택시"이용 금액만 월별 합계를 구하여, plot을 그린다.

taxi=[0,0,0]

for row in data:
    if row[-1]=='전표매입' and '택시' in row[5]:
        mon = int(row[0].split('-')[1])
        taxi[mon-10] = taxi[mon-10] + int(row[-3])
        
plt.title("10~12월 택시비 지출현황")
plt.plot(['10월','11월','12월'], taxi, color='crimson', label='택시비 지출액') 
plt.legend()
plt.show()
```

```python
## "전표매입"데이터 중 "택시"이용 금액과 "배달음식비용"의 월별 합계를 구하여, plot을 그린다.

taxi=[0,0,0]
for row in data:
    if row[-1]=='전표매입' and '택시' in row[5]:
        mon = int(row[0].split('-')[1])
        taxi[mon-10] = taxi[mon-10] + int(row[-3])


deli=[0,0,0]
for row in data:
    if row[-1]=='전표매입' and '(주)우아한형제들' in row[5]:
        mon = int(row[0].split('-')[1])
        deli[mon-10] = deli[mon-10] + int(row[-3])

plt.title("10~12월 택시비/배달음식비 지출현황")
plt.plot(['10월','11월','12월'], taxi, color='crimson', label='택시비 지출액')     
plt.plot(['10월','11월','12월'], deli, color='indigo', label='배달음식비 지출액')  
plt.legend()
plt.show()
```
